```python
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import sys
sys.path.append("..")
from mibi.preprocess.mibi_image import *
from mibi.preprocess.mibi_plot import *
```

```python
data_dir = '/Users/mschachter/data/CHOP_TMA_pilot'
images_dir = os.path.join(data_dir, 'extracted')
channels_data_path = os.path.join(data_dir, 'info', 'channels.csv')

mp_images = dict()
for point_name in ['Point9', 'Point16']:
    pdir = os.path.join(images_dir, point_name)
    _,pname = os.path.split(pdir)
    print(f"Loading from {pdir}")
    mp_images[pname] = MIBIMultiplexImage.load_from_path(os.path.join(pdir, 'TIFs'), channels_data_path)
```

```python
for pname,mp_img in mp_images.items():
    mp_img.bg_subtract()
    mp_img.threshold(percentile_threshold=75)
    mp_img.denoise()
```

```python
fig = plot_mibi_image(mp_images['Point9'], transforms_to_show=['raw', 'denoise'])
```

```python
out_imgs_dir = os.path.join(data_dir, 'processed')
mp_images['Point9'].write(os.path.join(out_imgs_dir, 'Point9'))
```

```python
fig = plot_mibi_image(mp_images['Point16'], transforms_to_show=['raw', 'denoise'])
```

```python
out_imgs_dir = os.path.join(data_dir, 'processed')
mp_images['Point16'].write(os.path.join(out_imgs_dir, 'Point16'))
```
